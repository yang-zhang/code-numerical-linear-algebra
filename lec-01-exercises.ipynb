### (a)

```python
import numpy as np
import pandas as pd
def df(x): return pd.DataFrame(x).round(2)
```

```python
m = 4
B = np.random.choice(20, (m, m), replace=False)
df(B)
```

#### 1. double col 1

```python
M1 = np.identity(4)
M1[0, 0] = 2
df(M1)
```

```python
B1 = B @ M1
df(B1)
```

#### 2. halve row 3
Use transpose to conver row problem to column problem.

halve row 3 of R <=> halve col 3 of R-transpose (RT)

A.T @ B.T = (B @ A).T

```python
B1T = B1.T
df(B1T)
```

```python
M2T = np.identity(m)
M2T[2, 2] = .5
df(M2T)
```

```python
B2T = (B1T @ M2T)
df(B2T)
```

```python
M2 = M2T.T
B2 = B2T.T
df(B2)
```

```python
assert np.allclose(B2, M2 @ B @ M1)
```

#### 3. add row 3 to row 1

```python
B2T = B2.T

M3T = np.identity(m)
M3T[2, 0] = 1
df(M3T)
```

```python
df(B2T)
```

```python
B3T = B2T @ M3T
df(B3T) 
```

```python
B3 = B3T.T
M3 = M3T.T
```

```python
df(B2)
```

```python
df(B3)
```

```python
assert np.allclose(B3, M3 @ M2 @ B @ M1)
```

#### 4. interchange cols 1 and 4

```python
M4 = np.identity(m)
```

```python
M4[0, 0] = 0
M4[3, 0] = 1
M4[3, 3] = 0
M4[0, 3] = 1
df(M4)
```

```python
df(B3)
```

```python
B4 = B3 @ M4
df(B4)
```

```python
assert np.allclose(B4, M3 @ M2 @ B @ M1 @ M4)
```

#### 5. subtract row 2 from each of the other rows

```python
B4T = B4.T
```

```python
M5T = np.identity(m)
```

```python
M5T[1, 0] = -1
M5T[1, 2] = -1
M5T[1, 3] = -1
df(M5T)
```

```python
df(B4T)
```

```python
B5T = B4T @ M5T
df(B5T)
```

```python
M5 = M5T.T
B5 = B5T.T
```

```python
df(B4)
```

```python
df(B5)
```

```python
assert np.allclose(B5, M5 @ M3 @ M2 @ B @ M1 @ M4)
```

#### 6. replace col 4 by col 3

```python
M6 = np.identity(m)
M6[3, 3] = 0
M6[2, 3] = 1
df(M6)
```

```python
B6 = B5 @ M6
df(B5)
```

```python
df(B6)
```

```python
assert np.allclose(B6, M5 @ M3 @ M2 @ B @ M1 @ M4 @ M6)
```

#### 7. delete column 1 

```python
M7 = np.zeros((m, m-1))
M7[1:, :] = np.identity(m-1)
df(M7)
```

```python
df(B6)
```

```python
B7 = B6 @ M7
df(B7)
```

```python
assert np.allclose(B7, M5 @ M3 @ M2 @ B @ M1 @ M4 @ M6 @ M7)
```
